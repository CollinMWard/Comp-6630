
The neural network has an input layer x, a hidden layer h with weights W1, biases b1, and a Sigmoid activation function, and an output layer with weights W2, biases b2, and linear activation for regression.

The forward pass involves computing z1 = W1x + b1.

Apply the Sigmoid activation function to get h = 1 / (1 + exp(-z1)).

Compute the output layer z2 = W2h + b2.

The prediction is given by y_hat = z2.

The loss function is the Mean Squared Error (MSE), defined as L = 0.5 * (y_hat - y)^2.

The gradient of the loss with respect to y_hat is dL/dy_hat = y_hat - y.

The gradient of the loss with respect to z2 is dL/dz2 = y_hat - y.

At the output layer, the gradient of the loss with respect to W2 is dL/dW2 = (y_hat - y) * h.

The gradient of the loss with respect to b2 is dL/db2 = y_hat - y.

For the hidden layer, the gradient of the loss with respect to h is dL/dh = (y_hat - y) * W2.

The gradient of the loss with respect to z1 is dL/dz1 = dL/dh * h * (1 - h) = (y_hat - y) * W2 * h * (1 - h).

The gradient of the loss with respect to W1 is dL/dW1 = dL/dz1 * x = [(y_hat - y) * W2 * h * (1 - h)] * x.

The gradient of the loss with respect to b1 is dL/db1 = dL/dz1 = (y_hat - y) * W2 * h * (1 - h).

The update rules using gradient descent are:

W2 = W2 - eta * dL/dW2.

b2 = b2 - eta * dL/db2.

W1 = W1 - eta * dL/dW1.

b1 = b1 - eta * dL/db1.

For binary classification with log loss, the loss function is L = -[y * log(y_hat) + (1 - y) * log(1 - y_hat)].

The output layer uses a Sigmoid activation function y_hat = 1 / (1 + exp(-z2)).

The gradient of the loss with respect to z2 becomes dL/dz2 = y_hat - y.

This introduces nonlinearity in the gradient flow due to the Sigmoid function at the output layer.

The primary difference between regression and binary classification is that regression uses a linear output activation and gradients directly scale with the error y_hat - y.

In binary classification, the Sigmoid activation modifies the gradient, making it suitable for probabilistic outputs.
This results in different update rules for the output layer between the two tasks.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_squared_error
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Dropout



X_train = np.loadtxt('/content/X_train.csv', delimiter=',', dtype=str)
Y_train = np.loadtxt('/content/Y_train.csv', delimiter=',', dtype=float)
X_test = np.loadtxt('/content/X_test.csv', delimiter=',', dtype=str)
Y_test = np.loadtxt('/content/Y_test.csv', delimiter=',', dtype=float)

X_train = np.array([list(map(float, x.split())) for x in X_train])
X_test = np.array([list(map(float, x.split())) for x in X_test])
Y_train = Y_train.reshape(-1, 1)
Y_test = Y_test.reshape(-1, 1)

#building model, activation = sigmoid is the hidden layer not output
def build_model(hidden_neurons, learning_rate, activation='sigmoid'):
    model = Sequential()
    model.add(Input(shape=(X_train.shape[1],)))
    # Add the hidden layer with the specified number of neurons
    model.add(Dense(hidden_neurons, activation=activation))
    model.add(Dense(1, activation='linear')) #output Layer
    model.compile(optimizer=Adam(learning_rate=learning_rate), loss='mse')
    return model

def calculate_r_squared(Y_true, Y_pred):
    ss_res = np.sum((Y_true - Y_pred) ** 2)  # Sum of squared residuals
    ss_tot = np.sum((Y_true - np.mean(Y_true)) ** 2)  # Total sum of squares
    r_squared = 1 - (ss_res / ss_tot)  # R-squared calculation
    return r_squared

hidden_neurons = 15
learning_rate = 0.1
model = build_model(hidden_neurons, learning_rate)
history = model.fit(X_train, Y_train, epochs=200, validation_split=0.2, batch_size=8, verbose=0)
Y_pred = model.predict(X_test)
mse_test = mean_squared_error(Y_test, Y_pred)
r_squared_value = calculate_r_squared(Y_test, Y_pred)

print(f'Mean Squared Error on Test Data: {mse_test}')
print(f'R-squared: {r_squared_value:.4f}')

# 4 Plot the loss over iterations
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

learning_rates = [0.001, 0.01, 0.1, 0.5, 1.0, 1.5, 2.0]
mse_values = []

for lr in learning_rates:
    model = build_model(hidden_neurons, lr)
    model.fit(X_train, Y_train, epochs=200, batch_size=8, verbose=0)
    Y_pred = model.predict(X_test)
    mse = mean_squared_error(Y_test, Y_pred)
    mse_values.append(mse)

plt.plot(learning_rates, mse_values, marker='o')
plt.title('Effect of Learning Rate on MSE')
plt.xlabel('Learning Rate')
plt.ylabel('MSE')
plt.xscale('log')
plt.show()

# Question 2
#Effect of number of neurons in the hidden layer
neuron_counts = range(1, 11)
mse_values_neurons = []
learning_rate = 0.1
all_predictions = []

for neurons in neuron_counts:
    model = build_model(neurons, learning_rate)
    history = model.fit(X_train, Y_train, epochs=200, batch_size=8, verbose=0, validation_split=0.2)
    Y_pred = model.predict(X_test)
    all_predictions.append(Y_pred)
    mse = mean_squared_error(Y_test, Y_pred)
    mse_values_neurons.append(mse)
    print(f'Number of Neurons: {neurons}, MSE: {mse:.4f}')

# Plot MSE for different numbers of neurons in the hidden layer
plt.plot(neuron_counts, mse_values_neurons, marker='o')
plt.title('Effect of Number of Neurons on MSE')
plt.xlabel('Number of Neurons')
plt.ylabel('Mean Squared Error')
plt.show()

# Plot the true vs. predicted labels for the model with the best number of neurons
best_neurons_index = np.argmin(mse_values_neurons)
best_neurons = neuron_counts[best_neurons_index]
best_predictions = all_predictions[best_neurons_index]

plt.scatter(Y_test, best_predictions, alpha=0.7, color='blue', label=f'{best_neurons} Neurons')
plt.xlabel('True Labels')
plt.ylabel('Predicted Labels')
plt.title(f'True vs. Predicted Labels (Best Model with {best_neurons} Neurons)')
plt.legend()
plt.show()
print(f'Best number of neurons in the hidden layer: {best_neurons} with MSE: {mse_values_neurons[best_neurons_index]:.4f}')

#7
activation_functions = ['relu', 'tanh']
mse_values_activation = []

for activation in activation_functions:
    model = build_model(hidden_neurons, learning_rate, activation=activation)
    history = model.fit(X_train, Y_train, epochs=200, batch_size=8, verbose=0, validation_split=0.2)
    Y_pred = model.predict(X_test)
    mse = mean_squared_error(Y_test, Y_pred)
    mse_values_activation.append(mse)

    # Plot true labels vs. predicted labels for each activation function
    plt.scatter(Y_test, Y_pred, label=f'Activation: {activation}', alpha=0.7)
    plt.xlabel('True Labels')
    plt.ylabel('Predicted Labels')
    plt.title(f'True vs. Predicted Labels ({activation})')
    plt.legend()
    plt.show()

# Bar chart comparing MSE for different activation functions
plt.bar(activation_functions, mse_values_activation, color=['blue', 'green'])
plt.title('Comparison of Activation Functions on MSE')
plt.xlabel('Activation Function')
plt.ylabel('Mean Squared Error')
for i, mse in enumerate(mse_values_activation):
    plt.text(i, mse, f'{mse:.4f}', ha='center', va='bottom')
plt.show()



# 8
# Train the model with early stopping

patience_values = [5, 10, 20]
history_records = {}

# Train the model with different patience values
for patience in patience_values:
    early_stop = EarlyStopping(monitor='val_loss', patience=patience, restore_best_weights=True)
    model = build_model(hidden_neurons=15, learning_rate=0.1)
    history = model.fit(X_train, Y_train, epochs=200, validation_split=0.2, batch_size=8,
                        verbose=0, callbacks=[early_stop])
    history_records[patience] = history.history
    Y_pred = model.predict(X_test)
    mse_test = mean_squared_error(Y_test, Y_pred)
    r_squared_value = calculate_r_squared(Y_test, Y_pred)
    print(f'Patience={patience}: MSE={mse_test:.4f}, R-squared={r_squared_value:.4f}')

    # Plot the training and validation loss
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'Training vs. Validation Loss (Patience={patience})')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()


#9 Dropout

def build_model_with_dropout(hidden_neurons, learning_rate, dropout_rate, activation='sigmoid'):
    model = Sequential()
    model.add(Input(shape=(X_train.shape[1],)))
    model.add(Dense(hidden_neurons, activation=activation))
    model.add(Dropout(dropout_rate))  # Apply variable dropout rate
    model.add(Dense(1, activation='linear'))
    model.compile(optimizer=Adam(learning_rate=learning_rate), loss='mse')
    return model

dropout_rates = [0.3, 0.2, 0.1]
history_records = {}

# Train and emodify dropout rates
for dropout_rate in dropout_rates:
    model_dropout = build_model_with_dropout(hidden_neurons=15, learning_rate=0.1, dropout_rate=dropout_rate)
    history_dropout = model_dropout.fit(X_train, Y_train, epochs=200, validation_split=0.2, batch_size=8, verbose=0)
    history_records[dropout_rate] = history_dropout.history
    Y_pred_dropout = model_dropout.predict(X_test)
    mse_test_dropout = mean_squared_error(Y_test, Y_pred_dropout)
    r_squared_value_dropout = calculate_r_squared(Y_test, Y_pred_dropout)
    print(f'Dropout Rate={dropout_rate*100:.0f}%: MSE={mse_test_dropout:.4f}, R-squared={r_squared_value_dropout:.4f}')

    # Plot the training and validation loss curves
    plt.figure(figsize=(8,6))
    plt.plot(history_dropout.history['loss'], label='Training Loss')
    plt.plot(history_dropout.history['val_loss'], label='Validation Loss')
    plt.title(f'Training vs. Validation Loss (Dropout={dropout_rate*100:.0f}%)')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

```
