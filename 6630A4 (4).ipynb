```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Load the data
file_path = '/content/Asssignment4_Data.xlsx'
train_data = pd.read_excel(file_path, sheet_name='Train')
test_data = pd.read_excel(file_path, sheet_name='Test')

X_train = train_data.drop(columns=['Construction type'])
y_train = train_data['Construction type']
X_test = test_data.drop(columns=['Construction type'])
y_test = test_data['Construction type']

clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

train_accuracy = accuracy_score(y_train, clf.predict(X_train))
test_accuracy = accuracy_score(y_test, clf.predict(X_test))

print(f"1 a. Accuracy on the training set: {train_accuracy:.2f}")
print(f" b. Accuracy on the test set: {test_accuracy:.2f}")
print(f"It is clearly overfitting the training data \n")

# Part 2:
depth_accuracies = []
for max_depth in range(1, 11):
    clf = DecisionTreeClassifier(max_depth=max_depth, random_state=42)
    clf.fit(X_train, y_train)
    test_acc = accuracy_score(y_test, clf.predict(X_test))
    depth_accuracies.append((max_depth, test_acc))
    print(f"Depth: {max_depth}, Test Accuracy: {test_acc:.2f}")

best_depth, best_accuracy = max(depth_accuracies, key=lambda x: x[1])
print(f"Best depth: {best_depth} with Test Accuracy: {best_accuracy:.2f}")

# Part 3:
print("""
3.
Restricting the depth of the decision tree prevents overfitting by limiting
the complexity of the model. A very deep tree will likely memorize the training
data, leading to high accuracy on the training set but poor generalization to
the test set. In this example, a depth of 2 is best balnce of complexity and simplicity
""")

# Part 4:
X_train = train_data.drop(columns=['Construction type', 'House ID'])
y_train = train_data['Construction type']
X_test = test_data.drop(columns=['Construction type', 'House ID'])
y_test = test_data['Construction type']

custom_test_point = pd.DataFrame([{
    "Local Price": 9.0384,
    "Bathrooms": 1,
    "Land Area": 7.8,
    "Living area": 1.5,
    "# Garages": 1.5,
    "# Rooms": 7,
    "# Bedrooms": 3,
    "Age of home": 23,
}])

clf = DecisionTreeClassifier(max_depth=best_depth, random_state=42)
clf.fit(X_train, y_train)

custom_prediction = clf.predict(custom_test_point)
print(f"4. Predicted class for the custom test point: {custom_prediction[0]}")
```
